```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("World_Alcohol.csv")
data.head(101)
```

```python
rows = data[~data['Display Value'].isnull()]
rows
```

```python
nullrows = data[data['Display Value'].isnull()]
nullrows
```

```python
data = data.dropna()
data = data.reset_index(drop=True)
data
```

```python
average = data['Display Value'].mean()
average
```

```python
data = data.replace(to_replace=0, value=average)
data
```

```python
data['Display Value'] = data['Display Value'].fillna(1.5)
data.head(30)
```
